Import statements

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load and display some of the data and basic statistics

```python
# loads and prints first couple rows of the data
data = pd.read_csv('heart.csv')
print(data.head())

# displays basic statistics
print(data.describe())
```

Histogram for the age

```python
plt.figure(figsize=(10, 6))
#plots the age distribution into 20 bins, along with a Kernel Density Estimate, which is a smooth estimate of the distribution
sns.histplot(data['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Box plot for resting blood pressure (trtbps) across sex

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='trtbps', data=data)
plt.title('Resting Blood Pressure by Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Resting Blood Pressure (mm Hg)')
plt.show()
```

Correlation heatmap

```python
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm') #data.corr returns the correlation matrix
plt.title('Correlation Heatmap')
plt.show()
```

Looks like the variables that are most correlated are:  

Positive:

1. max heart rate (thalachh) and overall likelihood of getting a heart attack (output)
2. max heart rate (thalachh) and slp (slope)


Negative:

1. slp (slope) and oldpeak ("previous peak")
2. exng (exercise-induced angina) and max heart rate (thalachh)
3. oldpeak ("previous peak") and max heart rate (thalachh)
4. exng (exercise-induced angina) and overall likelihood of getting a heart attack (output)
5. oldpeak ("previous peak") and overall likelihood of getting a heart attack (output)

Scatter plot for max heart rate vs age

```python
plt.figure(figsize=(10, 6))

# plots scatter plot for max heart rate vs age. Orange is less chance of a heart attack, while blue is more.
sns.scatterplot(x='age', y='thalachh', hue='output', data=data) 
plt.title('Maximum Heart Rate vs. Age')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (thalachh)')
plt.legend(title='Output', labels=['Less chance of heart attack', 'More chance of heart attack'])
plt.show()
```
